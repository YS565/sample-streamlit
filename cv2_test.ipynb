```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('C:/Users/koji/Desktop/image.jpg')
```

```python
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10, 8))

#imageの形状確認
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
p = cv2.Laplacian(img, cv2.CV_64F).var()
print("ボケの程度は%fとする" % p)
plt.imshow(img)
```

```python
#シャープ化フィルタリング
kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]], np.float32)

#以下の値に変更すると更にシャープになる( 次のセクションでやってみます)
#kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]], np.float32) 

res = cv2.filter2D(img,-1,kernel)
p2 = cv2.Laplacian(res, cv2.CV_64F).var()
s = "ボケの程度は「%f」から「%f」になり、シャープになったと言える" % (p,p2)
print(s)
plt.figure(figsize=(10, 8))
plt.imshow(res)
```

```python
#以下のシャープ化フィルタも試してみる
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]], np.float32) 

res = cv2.filter2D(img,-1,kernel)
p2 = cv2.Laplacian(res, cv2.CV_64F).var()
s = "ボケの程度は「%f」から「%f」になり、かなりシャープになったと言える" % (p,p2)
print(s)
plt.figure(figsize=(10, 8))
plt.imshow(res)
```

```python
cv2.imshow('img',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
